```python
import matplotlib.pyplot as plt
```

```python
cnn_accuracy = 0.98      # example: 98% (use your CNN result)
lstm_accuracy = 0.99     # from your LSTM notebook
```

```python
models = ['CNN', 'CNN + LSTM']
accuracies = [cnn_accuracy, lstm_accuracy]

plt.bar(models, accuracies)
plt.ylabel("Accuracy")
plt.title("CNN vs CNN+LSTM Accuracy Comparison")
plt.ylim(0.9, 1.0)
plt.show()
```

The CNN+LSTM model outperforms the CNN model by learning both spatial and temporal features.
Hence, CNN+LSTM is more suitable for driver drowsiness detection.
